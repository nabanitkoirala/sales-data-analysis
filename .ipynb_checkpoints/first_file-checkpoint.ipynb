```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
#to avoid error,use 'unicode_escape'
```

```python
df.shape
# shows the rows and columns in the file or data frame
```

```python
#deleting unnecessary/blank columns axis=1 means deleting whole row and inplace=True updates the table
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df)
#This shows all nullable and non nullable value in the table
```

```python
#check for null values and shows aggrigated sum of which row consists of how many null value
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
#change data types
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
#checking the data type of amount column
```

```python
df.columns
#checking columns in datasets
```

```python
#rename_column
df.rename(columns={'Marital_Status':'Marriage'})
```

```python
# describe() method returns description of the data in the Dataframe (i.e count,mean,std,etc)
df.describe()
```

```python
#use describe() for specific columns
df[['Age','Orders','Amount']].describe()
```

```python
df.columns
```

```python
sns.countplot(x = "Gender",data=df)
```

```python
ax=sns.countplot(x = "Gender",data=df)
#storing value of data in container
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
bx=sns.barplot(x='Gender',y='Amount',data=sales_gen)
for bars in bx.containers:
    bx.bar_label(bars,fmt='%.0f')
```

```python
sns.countplot(data=df,x='Age Group',hue='Gender')
#Hue gives a legend of gender and classify data according to gender
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total Amount vs Age Group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)

#From above graph we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
# total number of orders from top 10 states

sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
#sns.set helps to fix the size of the plot (width,height)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='State',y='Orders',hue='State')
```

```python
# total amount/sales from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
#sns.set helps to fix the size of the plot (width,height)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='State',y='Amount',hue='State')
```

```python
#From above graphs we can see that most of the orders and total sales/amount are from Uttar pradesh,maharastra and karnataka respectively
```

```python
#Maritial Status
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
#sns.set helps to fix the size of the plot (width,height)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

```python
#From above graph we can see that most of the buyers are married(women) and they have high purchasing power
```

```python
#Occupation
sns.set(rc={'figure.figsize':(23,5)})
ax=sns.countplot(data=df,x='Occupation',hue='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
#sns.set helps to fix the size of the plot (width,height)
sns.set(rc={'figure.figsize':(23,5)})
sns.barplot(data=sales_state,x='Occupation',y='Amount',hue='Amount')
```

```python
# From above graph we can see that most of the buyers are working in IT,Healthcare and Aviation sector
```

```python
#Product Category
sns.set(rc={'figure.figsize':(28,5)})
ax=sns.countplot(data=df,x='Product_Category',hue='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
#sns.set helps to fix the size of the plot (width,height)
sns.set(rc={'figure.figsize':(23,5)})
sns.barplot(data=sales_state,x='Product_Category',y='Amount',hue='Amount')
```

```python
#from above graph we can see that most of the sold products are from Food,Clothing and Electronics category
```

```python
sales_state=df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
#sns.set helps to fix the size of the plot (width,height)
sns.set(rc={'figure.figsize':(23,5)})
sns.barplot(data=sales_state,x='Product_ID',y='Amount',hue='Amount')
```

```python
#Conclusion:Married women age group 26-35 yrs from up,Maharastra and karnataka working in IT,Healthcare and Aviation are more likely to buy products from food,Clothing and Electronics category
```
